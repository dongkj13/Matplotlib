# Axes

## Sample

```python
x = np.arange(-np.pi, np.pi, 0.1)

fig, ax = plt.subplots(figsize = (8, 8))

# 绘图
ax.plot(x, np.cos(x), c = 'b', lw = 2.5, label = 'cosine')
ax.plot(x, np.sin(x), c = 'r', lw = 2.5, label = 'sine')

# 画线，文本标注
t = 2 * np.pi / 3
ax.vlines(x = t, ymin = np.cos(t), ymax = 0, colors = 'b', lw = 2.5, linestyles = "--")
ax.scatter([t, ], [np.cos(t), ], 50, color = 'b')
ax.annotate(r'$cos(\frac{2\pi}{3})=-\frac{1}{2}$', xy=(t, np.cos(t)), xycoords='data',\
 xytext=(-90, -50), textcoords='offset points', fontsize=15,
 arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"));

ax.vlines(x = t, ymin = 0, ymax = np.sin(t), colors = 'r', lw = 2.5, linestyle = "--")
ax.scatter([t, ], [np.sin(t), ], 50, color = 'r')
ax.annotate(r'$sin(\frac{2\pi}{3})=\frac{\sqrt{3}}{2}$', xy=(t, np.sin(t)), xycoords='data',\
 xytext=(+10, +30), textcoords='offset points', fontsize=16,
 arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"));

# 画矩形
ax.axvspan(-1, 1, alpha = 0.5, lw = 2, ls = '-.', ec = 'k', fc = 'pink')

# 所有画图类型都有label
ax.legend(labels = ['a', 'b', 'c', 'd','e','f','h','i','j'], loc = 'upper left', fontsize = 15)

# 对x轴主要刻度线设置网格，两种方式一样
ax.grid(True, which = 'major', axis = 'x', c = 'gray', ls = '--', alpha = 0.5)
#ax.xaxis.grid(True, which = 'major', c = 'gray', ls = '--', alpha = 0.5)  #这种方式在下一节讲到

# 坐标轴名称
ax.set_xlabel('x', fontsize = 20);
ax.set_ylabel('y', fontsize = 20, rotation = 0);

# 图像名称
ax.set_title('y = f(x)', fontsize = 20);

```

# Axis

## Sample

```python
from math import gcd
x = np.arange(-2 * np.pi, 2 *np.pi, 0.1)

fig, ax = plt.subplots(figsize = (8, 8))

# 绘图
ax.plot(x, np.cos(x), c = 'b', lw = 2.5, label = 'cosine')

def pi_formatter(x, pos):  
    """ 
    比较罗嗦地将数值转换为以pi/n为单位的刻度文本 
    """  
    n = 2  #以pi/n为单位
    m = int(np.round(x / (np.pi/n)))
    g = gcd(m, n)
    m, n = m // g, n // g
    if m == 0:  
        return "0"  
    if m < 0:
        s = "-"
    else:
        s = ""            
    m = np.abs(m)
    if m == 1 and n == 1:  
        return s + "$\pi$"
    if n == 1:  
        return s + (r"$%d \pi$" % m)
    if m == 1:  
        return s + (r"$\frac{\pi}{%d}$" % n)
    return s + (r"$\frac{%d \pi}{%d}$" % (m,n))


# 设置主、次刻度标签的位置及文本
ax.xaxis.set_major_formatter(mtick.FuncFormatter(pi_formatter))
ax.xaxis.set_major_locator(mtick.MultipleLocator(np.pi / 2))
ax.xaxis.set_minor_locator(mtick.MultipleLocator(np.pi / 10))
ax.xaxis.grid(True, which='major', ls = '--') #x坐标轴的网格使用主刻度
ax.xaxis.grid(False, which='minor') #x坐标轴的网格不使用次刻度
ax.xaxis.set_ticks_position('top')


# 刻度线参数设置
ax.xaxis.set_tick_params(which='both', width=1, labelsize = 15)
ax.xaxis.set_tick_params(which='major', length=7)
ax.xaxis.set_tick_params(which='minor', length=4)
```

## 主要参数

- [ax.xaxis.set_major_locator](https://matplotlib.org/api/ticker_api.html#tick-locating)：设置主刻度标签位置（[与日期有关的](https://matplotlib.org/api/dates_api.html#date-tickers)）
- [ax.xaxis.set_major_formatter](https://matplotlib.org/api/ticker_api.html#tick-formatting)：设置主刻度标签格式（[与日期有关的](https://matplotlib.org/api/dates_api.html#date-formatters)）
- [ax.xaxis.set_tick_params](https://matplotlib.org/api/_as_gen/matplotlib.axis.Axis.set_tick_params.html#matplotlib.axis.Axis.set_tick_params)：刻度线参数设置
- [ax.xaxis.set_ticks_position](https://matplotlib.org/api/_as_gen/matplotlib.axis.XAxis.set_ticks_position.html#matplotlib.axis.XAxis.set_ticks_position)：x轴刻度线位置

# Spine

## Sample

```python
x = np.arange(-np.pi, np.pi, 0.1)

fig, ax = plt.subplots(figsize = (8, 8))

ax.plot(x, np.sin(x), c = 'r', lw = 2.5, label = 'sine')

ax.spines['top'].set(visible = False);
ax.spines['right'].set(visible = False);
ax.spines['bottom'].set(color = 'b', lw = 3, position = ('data',0));
ax.spines['left'].set(color = 'g', lw = 3, position = ('data',0));
```

# Tick Locator

## Sample

```python
def tickline(ax):
    ax.set_xlim(0, 10), ax.set_ylim(-1, 1), ax.set_yticks([])
    ax.spines['right'].set_color('none')
    ax.spines['left'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.xaxis.set_ticks_position('bottom')
    ax.spines['bottom'].set_position(('data',0))
    ax.yaxis.set_ticks_position('none')
    ax.xaxis.set_minor_locator(mtick.MultipleLocator(0.1))
    ax.xaxis.set_tick_params(which='major', width = 2, length=7, labelsize = 15)
    ax.xaxis.set_tick_params(which='minor', width = 1, length=4)
    ax.plot(np.arange(11), np.zeros(11))

# 不同类型的locators
locators = [
    'mtick.NullLocator()',
    'mtick.MultipleLocator(1.0)',
    'mtick.FixedLocator([0, 2, 8, 9, 10])',
    'mtick.IndexLocator(3, 1)',
    'mtick.LinearLocator(5)',
    'mtick.LogLocator(2, [1, ])',
    'mtick.AutoLocator()',
]

n_locators = len(locators)

fig, axes = plt.subplots(ncols = 1, nrows = n_locators, figsize = (12, n_locators))

for i, locator in enumerate(locators):
    tickline(axes[i])
    axes[i].xaxis.set_major_locator(eval(locator))
    axes[i].text(5, 0.3, locator[6:], ha='center', fontsize = 20)
```

## Locator主要参数

- NullLocator：没有刻度标签
- [MultipleLocator](https://matplotlib.org/api/ticker_api.html#matplotlib.ticker.MultipleLocator)：在倍数处设置标签
- [FixedLocator](https://matplotlib.org/api/ticker_api.html#matplotlib.ticker.FixedLocator)：在给定处设置标签
- [IndexLocator](https://matplotlib.org/api/ticker_api.html#matplotlib.ticker.IndexLocator)：place ticks on the i-th data points where (i-offset)%base==0
- [LinearLocator](https://matplotlib.org/api/ticker_api.html#matplotlib.ticker.LinearLocator)：给定标签个数，均匀分布
- [LogLocator](https://matplotlib.org/api/ticker_api.html#matplotlib.ticker.LogLocator)：Place ticks on the locations : subs[j] * base**i
- [AutoLocator](https://matplotlib.org/api/ticker_api.html#matplotlib.ticker.AutoLocator)：自动设置
